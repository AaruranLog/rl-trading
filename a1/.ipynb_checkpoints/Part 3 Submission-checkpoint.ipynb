By increasing the update period, and allowing the target network to only update at larger intervals, the average cumulative rewards across the number of iterations increases more quickly. This is because the Q-network is able to optimize more thoroughly, before it's objective function is perturbed and corrected by the target's update. 

In value iteration, the value matrix $V$ is updated in the following way at every time step.
$$
    V \leftarrow \max_a R^a + \gamma T^a V
$$
In contrast, the target network weights are updated at the end of each minibatch with the weights of the Q-network. Value iteration is connected to the deep Q-network approach, in the sense that we can use a Q-table to recover the value  matrix.

```python
Image(filename='./minibatch-size-plot.png')
```

The larger the minibatch size, the more quickly the Q-network was able to train. There is some stochasticity inherent in each individual experience tuple, but by increasing the batch size, each network update has a comparatively reduced variance. This reduced variance improves the rate of convergence. 

This is analogous to minibatch gradient descent and exact gradient descent. Just as larger batch size  improves convergence in gradient descent, so too does it improve convergence via the replay buffer.
